```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler #for standardizing the data
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn import svm # support vector machine also logistic regree can be used but for small datasets with complex features
#we use svm it hard classifies things with a margin line how far the point is whether its true or not true thats it
```

```python
diabetes_dataset=pd.read_csv('/content/diabetes (1).csv')
```

```python
diabetes_dataset.head()
```

```python
# after importing check the outcome categories have equal counts like diabetic 50% and non diabetic also be like 50%
diabetes_dataset['Outcome'].value_counts()
```

0->non diabetic
1->diabetic

getting mean values of all features mean grouped by labels outcomes for cool understanding

```python
diabetes_dataset.groupby('Outcome').mean()
```

```python
#separating features and labels
X=diabetes_dataset.drop(columns='Outcome',axis=1)
Y=diabetes_dataset['Outcome']
X
```

Standardizing the data cuz all features are in different ranges
using StandardScaler funtion

```python
scaler=StandardScaler()
scaler.fit(X)
```

transforming the standardized data

```python
standardized_data=scaler.transform(X)
#or  we can fit and transform data at a time by "scaler.fit_transform()"
```

```python
print(standardized_data)
```

```python
X=standardized_data
Y=diabetes_dataset['Outcome']
```

train test split

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
# stratify funtion is used on y because
#this funtion used to spread the data in eqally
#like for y outcomes 1,0 should be spreded in x_train and x_test
#and x_test proposionally
```

```python
X.shape,X_train.shape,X_test.shape
```

Traning model with support vector machine svm model

```python
classifier=svm.SVC(kernel='linear') #loading the machine
```

```python
#training the svm classifier
```

```python
classifier.fit(X_train,Y_train)
```

model evaluation

```python
X_train_prediction=classifier.predict(X_train)
```

```python
x_train_accuracy=accuracy_score(X_train_prediction,Y_train)
```

```python
x_train_accuracy
```

```python
X_test_prediction=classifier.predict(X_test)
X_test_accuracy=accuracy_score(X_test_prediction,Y_test)
```

```python
X_test_accuracy
```

Making a predictive system

```python
input_data=(3,111,56,39,0,30.1,0.557,30)
# changing the input data to numpy array
input_data_as_np=np.asarray(input_data)
input_reshaped=input_data_as_np.reshape(1,-1)
# now lets standardize the data
std_data=scaler.transform(input_reshaped)
print(std_data)

prediction=classifier.predict(std_data)
print(prediction)
```
